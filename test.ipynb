```python
import requests
response = requests.get('https://quay.io/api/v1/repository/fedora/fedora/tag/')
tags = response.json()['tags']
```

```python
tag_names = ['latest', 'rawhide']
# Get the manifest for each tag
manifests = {}
for tag_name in tag_names:
    url = f"https://quay.io/api/v1/repository/fedora/fedora/tag/{tag_name}/manifest"
    response = requests.get("https://quay.io/api/v1/repository/fedora/fedora/tag/")
    manifest = response.json()
    print(manifest)
    manifests[tag_name] = manifest['manifest_digest']

# Get the list of tags that share the same manifest as the original tags
url = f"https://quay.io/api/v1/repository/fedora/fedora/manifest/{manifests['latest']}/tags"
response = requests.get(url)
tags = response.json()

# Print the list of tags
print(f"Tags that share a manifest with 'latest' and 'rawhide': {tags}")
```
